```python
import os
import pandas as pd
import matplotlib.pyplot as plt

def create_histograms_from_csv(csv_path):
    """
    Reads a CSV file and generates a histogram for each column.
    
    The title of each histogram will contain the file name and the column name.
    Histograms are saved as PNG files in the specified output directory.
    """
    
    # Extract the base filename from the csv_path (without extension)
    file_name = os.path.splitext(os.path.basename(csv_path))[0]
    
    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv_path)
    
    # Loop through each column in the DataFrame
    for col in range(len(df.columns)):
        plt.figure(figsize=(8, 6))
        # Create histogram; assuming all values are numeric and continuous
        plt.hist(df.iloc[:, col].dropna(), bins=30, edgecolor='black')
        # Set title with file name and column name
        plt.title(f"{file_name} - column:{col}")
        plt.xlabel(col)
        plt.ylabel("Frequency")
        
        # Save the histogram to a file in the output directory
        plt.show()
```

```python
for fitness_file_path in fitness_files:
    create_histograms_from_csv(fitness_file_path)
```
