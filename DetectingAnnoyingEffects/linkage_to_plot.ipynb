```python


import numpy as np

import utils

n = 20
clique_size = 5
linkage_table = np.zeros((n, n), dtype=float)

for clique_index in range(n//clique_size):
    start = clique_index * clique_size
    end = start + clique_size
    
    for row in range(start, end):
        for col in range(start, end):
            linkage_table[row, col] = 0.8
    
utils.show_matrix_with_colours(linkage_table, fig_size=(5, 5))
```

```python
import sklearn.manifold as skm

# this can be used for both exercises and students
def get_fitted_points(distance_matrix, result_vector_size, metric=True):
    mds_object = skm.MDS(n_components=result_vector_size, metric= metric, dissimilarity='precomputed')
    fit_transform_result = mds_object.fit_transform(distance_matrix)

    return fit_transform_result


def get_fitted_points_from_linkage_table(input_linkage_table):
    distance_table = 1 - input_linkage_table
    return get_fitted_points(distance_table, result_vector_size=2)
```

```python
points = get_fitted_points_from_linkage_table(linkage_table)
xs, ys = points[:, 0], points[:, 1]
utils.simple_scatterplot(x_label = "who knows", y_label="who knows", xs = xs, ys=ys)
```
